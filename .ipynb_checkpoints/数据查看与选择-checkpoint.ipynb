```python
import pandas as pd
```

```python
df = pd.read_excel('./team.xlsx')
```

```python
# 数据查看
df.head()
df.tail(50)
df.sample(3)
```

```python
# 数据验证
df.shape
df.info() # 查看索引 数据类型  内存信息
df.describe() # 数值类型列的汇总统计
df.dtypes
df.axes
df.columns
```

```python
# 索引设置
df.set_index('name',inplace=True)
df
```

```python
# 选取列 直接通过列明或者loc来
df['team'] # 选择单列
df.team
df[['team','Q1']] # 选择多列
df.loc[:,['team','Q2']] 
```

```python
# 选取行  直接通过iloc或者直接通过索引
df.iloc[0:10:3,0:2]  # 可以选择行，也能选择列
# df[df.index == 'Liver']
df[0:6] # 这种只能选择行，不能再选择列
```

```python
# 选择行列总结：可以通过下面两种方式
# 1.直接选取：列名切片，行索引切片   
# 2.方法选取：iloc，loc  这两种方式都可以通过连续和非连续两种取数的方式 
df.iloc[2:5,2:4]
df.iloc[[2,5],[2,4]]
df.loc['Liver':'Oah', 'Q2':'Q4'] # !!!
df.loc[['Liver','Eorge'], 'Q2':'Q4' ]
df.loc[['Liver','Eorge'], ['Q2', 'Q4'] ]
```

```python
# 总结以上选取数据的方式来说，要掌握iloc，loc 各两种共四种方式，以及直接选择的两种方式，掌握这六种常见的选择方式即可。
```

```python
# 条件选择
# 单一条件
df[df.index == 'Eorge']
df[df['Q1'] > 90]
df[df.team == 'B']

# 复合条件
df[(df.Q1 > 90) & (df.Q4 > 50)] # 注意条件要加括号，避免歧义
# df[df.Q1 > 90][df.Q2 > 50] # 不能这样写，前后选取的数据 索引不一致，长度也不一致
df[df.Q1 > 90].loc[df.Q1 > 50]
```

```python
# 排序
df.sort_values(by='Q1', ascending=False)
df.sort_values(by=['team','Q4'], ascending=[True,False])
```

```python
# 分组聚合
# 我们可以实现类似SQL的groupby那样的数据透视功能
df.groupby('team').sum()
df.groupby('team').count()
df.groupby('team').mean() # 求平均

# 不同的列聚合后用不同的计算方式

df.groupby('team').agg({
    'Q1' : 'sum',
    'Q2' : 'mean',  # 需要加‘’
    'Q3' : 'count',  # 需要加‘’
    'Q4' : 'max',
    
})

# 排序之后不展示team了，需要将结果附着到原来frame上，并且注意['Q1'] 不能对Q1 Q2全部排名
df['rank'] = df.groupby('team')['Q1'].rank() # 排序 组内排序  ？？  为啥不显示team了？ 没写按照谁来排名的
df
```

```python
# 数据转换
df.groupby('team').sum()
```

```python
# 数据转置
df.groupby('team').sum().T
df.groupby('team').sum().stack() # 分组堆叠展示 以所以为组
# df.groupby('team').sum().unstack()  # 分组堆叠展示 以列为组
```

```python
# 增加列
df['one'] = 1 # 增加一个固定值的列
df['total'] = df.Q1 + df.Q2 + df.Q3 + df.Q4 # 增加总成绩列
# # 将计算得来的结果赋值给新列
df['total2'] = df.loc[:,'Q1':'Q4'].apply(lambda x:sum(x), axis=1) #  ！！！ 对这一行的数据分别做不同的处理 axis=1是横向相加，=0是竖向相加
df['total3'] = df.sum(axis=1, numeric_only=True) # 横向0纵向1 默认0 index (0)默认, columns (1) 可以把所有为数字的列相加  ！！！  需要加参数numeric_only=True 
df['avg'] = df.total/4 # 增加平均成绩列
df
```

```python
df[0:2].loc[:,'Q1':'Q4'].apply(lambda x:sum(x), axis=0) # 竖向相加
```

```python
# 统计分析
df.mean(axis=0,numeric_only=True) # 返回所有列的均值
df.mean(axis=1,numeric_only=True) # 返回所有行的均值，下同
df.loc[:,'Q1':'Q4'].corr() # 返回列与列之间的相关系数
df.count(axis=1,numeric_only=True) # 返回每一列中的非空值的个数 axis=1可以得到横向的
df.max(axis=1,numeric_only=True) # 返回每一列的最大值
df.min(numeric_only=True) # 返回每一列的最小值
df.median(numeric_only=True) # 返回每一列的中位数
df.std(numeric_only=True) # 返回每一列的标准差
df.var(numeric_only=True) # 方差
```
