```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(np.random.randn(30,4), columns=['A','B','C','D'], index=pd.date_range('20221001', periods=30))
df
```

```python
df.rolling('3D', min_periods=3).mean()
```

```python
# 可以使用apply函数来对窗口的数据进行计算  注意apply是直接计算一次的，x应该是可迭代对象 sum全局函数
df.rolling(3).apply(lambda x: sum(x)/3 )
```

```python
# expanding使用: 使用expanding模拟cumprod, 效果和df.cumprod()是一样的
df.expanding(1).apply(lambda x: np.prod(x))
```

```python
df.cumprod()
```
